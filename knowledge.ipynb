# Knowledge Inference Notebook
This notebook contains a small forward-chaining inference engine. Edit the base facts and rules, then run the code cell to compute inferred facts.

```python
# Regenerated inference engine
from typing import List, Tuple, Set

Fact = Tuple[str, str, str]  # (subject, relation, object)

# Base facts - edit these as needed
facts: List[Fact] = [
    ('mango', 'is_a', 'fruit'),
]

def rule_fruit_implies_food(facts_set: Set[Fact]) -> Set[Fact]:
    """If (X, 'is_a', 'fruit') then infer (X, 'is_a', 'food')
    """
    new = set()
    for (s, r, o) in facts_set:
        if r == 'is_a' and o == 'fruit':
            inferred = (s, 'is_a', 'food')
            if inferred not in facts_set:
                new.add(inferred)
    return new

def rule_vegetable_implies_food(facts_set: Set[Fact]) -> Set[Fact]:
    """If (X, 'is_a', 'vegetable') then infer (X, 'is_a', 'food')
    """
    new = set()
    for (s, r, o) in facts_set:
        if r == 'is_a' and o == 'vegetable':
            inferred = (s, 'is_a', 'food')
            if inferred not in facts_set:
                new.add(inferred)
    return new

def apply_rules(facts: List[Fact]) -> List[Fact]:
    """Compute closure under rules and return sorted facts.
    """
    facts_set: Set[Fact] = set(facts)
    rules = [rule_fruit_implies_food, rule_vegetable_implies_food]
    changed = True
    while changed:
        changed = False
        new_facts = set()
        for rule in rules:
            new_facts |= rule(facts_set)
        for nf in new_facts:
            if nf not in facts_set:
                print('New fact added:', nf)
                facts_set.add(nf)
                changed = True
    return sorted(facts_set)

def query(facts_list: List[Fact], subject: str = None, relation: str = None, obj: str = None) -> List[Fact]:
    """Return facts matching the pattern (None = wildcard).
    """
    results = []
    for (s, r, o) in facts_list:
        if (subject is None or subject == s) and (relation is None or relation == r) and (obj is None or obj == o):
            results.append((s, r, o))
    return results

if __name__ == '__main__':
    print('Initial facts:')
    print(facts)

    print('\nRunning inference...')
    inferred = apply_rules(facts)
    print('Inference complete.\n')
    print('All inferred facts:')
    for f in inferred:
        print(' -', f)

    foods = [s for (s, r, o) in query(inferred, relation='is_a', obj='food')]
    print('\nQuery: What is \'food\'? ->', sorted(foods))
```
